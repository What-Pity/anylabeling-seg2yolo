## 读取json标注

```python
import json
import pprint as pp

with open('bus.json', 'r') as f:
    data = json.load(f)
    pp.pprint(data)
```

可以看到有宽、高、图片位置和标注坐标信息

看看标注信息

```python
pp.pprint(data["shapes"])
len(data["shapes"]),type(data["shapes"])
```

标注信息是NX4的list

```python
print(type(data["shapes"][0]),len(data["shapes"][0]),data["shapes"][0].keys(),sep="  ")
pp.pprint(data["shapes"][0])
```

标注信息中，重要的内容有label和points

```python
type(data["shapes"][0]["points"]),data["shapes"][0]["points"]
```

points里含有MX2的list

接下来取出所有有用信息

```python
import numpy as np
h=data["imageHeight"]
w=data["imageWidth"]
path=data["imagePath"]
shapes=data["shapes"]

sample={shapes[0]["label"]:np.array(shapes[0]["points"])}

print(h,w,path,sample,sep="\n")
```

归一化坐标

```python
for xy in sample.values():
    xy /= np.array([w,h])
print(sample.values())
```

看看对不对

```python
from matplotlib import pyplot as plt
%matplotlib inline
import PIL.Image as Image

image=Image.open("bus.jpg")
fig,axes=plt.subplots(1,2,figsize=(10,5))
axes[0].imshow(image)
axes[0].axis('off')
axes[1].plot(sample["human"][:,0],-sample["human"][:,1],'r.')
axes[1].axis('equal')
```
